```python
#Advanced 5 Find top 3 correlated attributes to survive 
df = pd.read_csv('train_titanic.csv')
df = df.replace(to_replace ="female", 
                 value =1)
df = df.replace(to_replace ="male", 
                 value =0) 
df.head()
df_corr = df.corr()['Survived'][2:] 
KeyToSurvive_list = abs(df_corr).sort_values(ascending=False)
KeyToSurvive_list=KeyToSurvive_list.head(3)
print("The top {} correlated values with Survived:\n{}".format(len(KeyToSurvive_list), KeyToSurvive_list))
```
